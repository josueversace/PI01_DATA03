## Leyendo Laptimes

#### Leemos de uno por uno y luego concatenamos con pd.concat

```python
union_lap1 = 'Datasets/lap_times/lap_times_split_1.csv'
union_lap01 = pd.read_csv(union_lap1, header=None)

union_lap2= 'Datasets/lap_times/lap_times_split_2.csv'
union_lap02 = pd.read_csv(union_lap2, header=None)

union_lap3 = 'Datasets/lap_times/lap_times_split_3.csv'
union_lap03 = pd.read_csv(union_lap3, header=None)

union_lap4 = 'Datasets/lap_times/lap_times_split_4.csv'
union_lap04 = pd.read_csv(union_lap4, header=None)

union_lap5 = 'Datasets/lap_times/lap_times_split_5.csv'
union_lap05 = pd.read_csv(union_lap5, header=None)

lap_times_concat = pd.concat([union_lap01, union_lap02, union_lap03, union_lap04, union_lap05], axis=0)
lap_times_concat2 = lap_times_concat.reset_index(drop=True)

lap_times_concat2
```

#### Corroboramos la integridad y datos faltantes

```python
lap_times_concat2.info()
```

#### Cambiando nombre de las columnas para poder manipularlas luego
4 = Tiempo 
1 = Driver id

```python
lap_times_concat2 = lap_times_concat2.rename(
    columns={lap_times_concat2.columns[4]: 'tiempo'})
lap_times_concat2 = lap_times_concat2.rename(
    columns={lap_times_concat2.columns[1]: 'Driver id'})
lap_times_concat2 = lap_times_concat2.rename(
    columns={lap_times_concat2.columns[0]: 'race id'})
    
lap_times_concat2
```

```python
lap_times_concat2.columns
```

```python
lap_times_concat2.info()
```

```python
lap_times_concat2['tiempo'].min()
```

#### Ordenando por tiempo

```python
by_time = lap_times_concat2.sort_values('tiempo')
by_time.head()
```

```python
by_time.tail()
```

#### Reseteamos el index

```python
lap_times_concat3 = by_time.reset_index(drop=True)
lap_times_concat3
```

## Una vez tenemos todos las tablas solo para elaborar consultas segun la consigna, Exportamos los datasets a una base de datos

## Usamos la libreria de sql alchemy

```python
# Nos conectamos a la base de datos
database_username = 'root'
database_password = '70040691'
database_ip = '127.0.0.1'
database_name = 'PI'
database_connection = sqlalchemy.create_engine('mysql+mysqlconnector://{0}:{1}@{2}/{3}'.
                                               format(database_username, database_password,
                                                      database_ip, database_name), pool_recycle=1, pool_timeout=57600).connect()
#Exportamos las tablas

# df Drivers
drivers_final.to_sql(con=database_connection, name='drivers',
            if_exists='replace', chunksize=100)
print('DF drivers cargado!')
print('1/9 COMPLETE!')

# df circuits
circuits.to_sql(con=database_connection, name='circuits',
                     if_exists='replace', chunksize=100)
print('DF circuits cargado!')
print('2/9 COMPLETE!')

# df races
races.to_sql(con=database_connection, name='races',
                     if_exists='replace', chunksize=100)
print('DF races cargado!')
print('3/9 COMPLETE!')

#result
result.to_sql(con=database_connection, name='results',
                     if_exists='replace', chunksize=100)
print('DF results cargado!')
print('4/9 COMPLETE!')

#constructores
const.to_sql(con=database_connection, name='constructors',
                     if_exists='replace', chunksize=100)
print('DF constructors cargado!')
print('5/9 COMPLETE!')


# qualifyings
qualify02.to_sql(con=database_connection, name='qualifying',
                     if_exists='replace', chunksize=100)
print('DF qualifying cargado!')
print('6/9 COMPLETE!')

# Pit stop
pit_stop.to_sql(con=database_connection, name='pit_stops',
                     if_exists='replace', chunksize=100)
print('DF pit_stops cargado!')
print('7/9 COMPLETE!')

# lap_times
lap_times_concat3.to_sql(con=database_connection, name='lap_times',
                     if_exists='replace', chunksize=100)
print('DF lap_times cargado!')
print('8/9 COMPLETE!')

#
result.to_sql(con=database_connection, name='results',
                     if_exists='replace', chunksize=100)
print('DF result cargado!')
print('9/9 COMPLETE!')

# Cerramos la sesion de nuestra base de datos
database_connection.close()
```
